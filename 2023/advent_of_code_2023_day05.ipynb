```python
# messing around w input data
# input_data = input_data.split("\n\n")

# getting the seeds
seeds = input_data[0].split(":")[1].strip().split(" ")
# getting all the maps
seed_soil = input_data[1].split(":")[1].strip().split("\n")
soil_fert = input_data[2].split(":")[1].strip().split("\n")
fert_water = input_data[3].split(":")[1].strip().split("\n")
water_light = input_data[4].split(":")[1].strip().split("\n")
light_temp = input_data[5].split(":")[1].strip().split("\n")
temp_humid = input_data[6].split(":")[1].strip().split("\n")
humid_loc = input_data[7].split(":")[1].strip().split("\n")

### into one block so that it can be run without running two different blocks

# getting the maps into a more workable list
def map_to_list(maps): # okay this works
  maps_list = []
  for map_str in maps:
    map_list = map_str.split(" ")
    map_origin_start = int(map_list[1])
    map_destin_start = int(map_list[0])
    map_range = int(map_list[2])

    maps_list.append([map_origin_start, map_destin_start, map_range])

  return maps_list

# applying the function
seed_soil = map_to_list(seed_soil)
soil_fert = map_to_list(soil_fert)
fert_water = map_to_list(fert_water)
water_light = map_to_list(water_light)
light_temp = map_to_list(light_temp)
temp_humid = map_to_list(temp_humid)
humid_loc = map_to_list(humid_loc)

# function from map_list to getting the result
def origin_destin(origin, map_list): # origin and destin should be int
  for map in map_list:
    origin_start = map[0]
    destin_start = map[1]
    map_range = map[2]

    if origin_start <= origin <= origin_start + map_range:
      destin = destin_start + (origin - origin_start)
      break
  else:
    destin = origin

  return destin

# funtion from seed to loc
def seed_to_loc(seed, seed_soil, soil_fert, fert_water, water_light, light_temp, temp_humid, humid_loc):
  # seed should be an int
  soil = origin_destin(seed, seed_soil)
  fert = origin_destin(soil, soil_fert)
  water = origin_destin(fert, fert_water)
  light = origin_destin(water, water_light)
  temp = origin_destin(light, light_temp)
  humid = origin_destin(temp, temp_humid)
  loc = origin_destin(humid, humid_loc)
  return loc
```

```python
# solving the problem
current_min = seed_to_loc(int(seeds[0]), seed_soil, soil_fert, fert_water, water_light, light_temp, temp_humid, humid_loc) # initialize with the first seed

for seed in seeds:
  seed = int(seed) # make them into an int first
  loc = seed_to_loc(seed, seed_soil, soil_fert, fert_water, water_light, light_temp, temp_humid, humid_loc)
  current_min = min(current_min, loc)

print(current_min)
# 35, right for the testcase
# 389056265, right first try
# took me 61' lmao
```

```python
# think in ranges (to not check all seeds one by one)
# man i should think about this from first principles

# i think for this i should make the seeds list into a list of int first
seeds = input_data[0].split(":")[1].strip().split(" ") # put this here so no need to run the upper blocks
# making them into a list of tuples of (seed_start, seed_end); seend_end is just seed_start + seed_range - 1
seeds = [(int(seeds[i]), int(seeds[i]) + int(seeds[i+1])-1) for i in range(0, len(seeds), 2)]

# redefine the origin_destin function
def origin_destin2(primary_ranges, map_list):
  # intersecting = [] # intersecting wih some map in map_list
  # non_intersecting = [] # not intersecting with some map, just maps to itself

  destin_ranges_list = [] # to store the list of destin ranges

  for primary_range in sorted(primary_ranges, key=lambda x: x[0]):
    # start = primary_range[0]
    # end = start + primary_range[1] # in the old seeds format
    start = primary_range[0]
    end = primary_range[1]

    for map in sorted(map_list, key=lambda x: x[0]): # sort the map_list by origin_start
      # some initialization
      map_range = map[2]
      #
      map_origin_start = map[0]
      map_origin_end = map_origin_start + map_range - 1 # don't forget to min 1
      #
      map_destin_start = map[1]
      #
      map_origin_to_destin = map_destin_start - map_origin_start # the translation


      # check if there's intersection with the range [map_origin_start, map_origin_end]
      # man this algo is genius holy shite
      if map_origin_start <= end and map_origin_end >= start:
        # establish the intersection
        intersection_origin_start = max(map_origin_start, start)
        intersection_origin_end = min(map_origin_end, end)

        # change the origin into the destin
        # add to destin_ranges_list
        destin_ranges_list.append((intersection_origin_start + map_origin_to_destin, intersection_origin_end + map_origin_to_destin))

        if start < map_origin_start:
          # if there's start before intersection
          # this falls outside the origin of the map, so it just maps to itself
          destin_ranges_list.append((start, map_origin_start - 1))

        # go from after the intersection
        start = intersection_origin_end + 1

    # if there's still a range left behind
    if start <= end:
      destin_ranges_list.append((start, end))

  return destin_ranges_list


def seed_to_loc2(seeds, seed_soil, soil_fert, fert_water, water_light, light_temp, temp_humid, humid_loc):
  # seed should be an int
  soil = origin_destin2(seeds, seed_soil)
  fert = origin_destin2(soil, soil_fert)
  water = origin_destin2(fert, fert_water)
  light = origin_destin2(water, water_light)
  temp = origin_destin2(light, light_temp)
  humid = origin_destin2(temp, temp_humid)
  loc = origin_destin2(humid, humid_loc)
  return loc

```

```python
# original chatgpt finding intersection logic, love it
def find_intersections_and_differences(primary_range, other_ranges):
    primary_start, primary_end = primary_range
    intersecting = []
    non_intersecting = []

    for other_range in sorted(other_ranges):
        other_start, other_end = other_range

        # Check if there is an intersection
        if other_start <= primary_end and other_end >= primary_start:
            intersection_start = max(primary_start, other_start)
            intersection_end = min(primary_end, other_end)
            intersecting.append((intersection_start, intersection_end))

            # Update primary range segments before and after the intersection
            if primary_start < intersection_start:
                non_intersecting.append((primary_start, intersection_start - 1))
            primary_start = intersection_end + 1

    # Add the remaining part of the primary range, if any
    if primary_start <= primary_end:
        non_intersecting.append((primary_start, primary_end))

    return non_intersecting, intersecting

# Example usage
primary_range = (26, 100)
other_ranges = [(20, 40), (90, 110)]

non_intersecting, intersecting = find_intersections_and_differences(primary_range, other_ranges)
print("Non-Intersecting Ranges:", non_intersecting)
print("Intersecting Ranges:", intersecting)
```
